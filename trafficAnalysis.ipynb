```python
import pandas as pd

df = pd.read_csv("traffic.csv")

print(df.head())
```

```python
print(df.info())
```

```python
#Vehicle Distribution Analysis:

import matplotlib.pyplot as plt
import seaborn as sns

vehicle_types = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']

plt.figure(figsize=(10, 5))
sns.boxplot(data=df[vehicle_types], palette="Set2")
plt.title("Boxplot of Vehicle Counts")
plt.xlabel("Vehicle Type")
plt.ylabel("Count")
plt.show()


plt.figure(figsize=(7, 20))
for i, vehicle in enumerate(vehicle_types, 1):
    plt.subplot(4, 1, i)
    sns.histplot(df[vehicle], bins=30, kde=True, color="green")
    plt.title(f"{vehicle} Distribution")
    plt.xlabel("Count")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Traffic Situation Distribution:

traffic_counts = df["Traffic Situation"].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=traffic_counts.index, y=traffic_counts.values, palette="pastel")
plt.title("Traffic Situation Distribution")
plt.xlabel("Traffic Situation")
plt.ylabel("Frequency")
plt.show()
```

```python
#Variation by Day of the Week:

day_counts = df.groupby("Day of the week")[vehicle_types].sum()

day_counts.plot(kind='bar', figsize=(10, 5), colormap="viridis")
plt.title("Vehicle Count Variation by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Count")
plt.legend(title="Vehicle Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#Car Count vs. Traffic Situation:

plt.figure(figsize=(10, 5))
sns.boxplot(x="Traffic Situation", y="CarCount", data=df, palette="coolwarm")
plt.title("Car Count vs. Traffic Situation")
plt.xlabel("Traffic Situation")
plt.ylabel("Car Count")
plt.show()
```

```python
#Bike Count vs. Traffic Situation:

plt.figure(figsize=(10, 5))
sns.boxplot(x="Traffic Situation", y="BikeCount", data=df, palette="coolwarm")
plt.title("Bike Count vs. Traffic Situation")
plt.xlabel("Traffic Situation")
plt.ylabel("Bike Count")
plt.show()
```

```python
#Bus Count vs. Traffic Situation:

plt.figure(figsize=(10, 5))
sns.boxplot(x="Traffic Situation", y="BusCount", data=df, palette="coolwarm")
plt.title("Bus Count vs. Traffic Situation")
plt.xlabel("Traffic Situation")
plt.ylabel("Bus Count")
plt.show()
```

```python
#Truck Count vs. Traffic Situation:

plt.figure(figsize=(10, 5))
sns.boxplot(x="Traffic Situation", y="TruckCount", data=df, palette="coolwarm")
plt.title("Truck Count vs. Traffic Situation")
plt.xlabel("Traffic Situation")
plt.ylabel("Truck Count")
plt.show()
```

```python
#Total Vehicle and Count Traffic Situation:

plt.figure(figsize=(10, 5))
sns.boxplot(x="Traffic Situation", y="Total", data=df, palette="coolwarm")
plt.title("Total Vehicle Count vs. Traffic Situation")
plt.xlabel("Traffic Situation")
plt.ylabel("Total Vehicle Count")
plt.show()
```

```python
#Busiest Hours of the Day:

df["Hour"] = pd.to_datetime(df["Time"], format="%I:%M:%S %p").dt.hour

hourly_counts = df.groupby("Hour")["Total"].sum()

plt.figure(figsize=(10, 5))
sns.barplot(x=hourly_counts.index, y=hourly_counts.values, palette="coolwarm")
plt.title("Busiest Hours of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Vehicle Count")
plt.xticks(range(0, 24))
plt.show()
```
